```python
cod_cmap= dados_cmap['Target'].value_counts().index[0:10]
```

```python
dados_cmap.query('Target in @cod_cmap')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(35, 12))
ax = sns.countplot(x = 'Target', data=dados_cmap.query('Target in @cod_cmap'), order= dados_cmap['Target'].value_counts().index[0:10], palette='husl')
ax.set_title('Top genes')
plt.title("Top genes", fontsize = 20)
plt.xlabel("genes", fontsize = 18)
plt.ylabel("drugs", fontsize = 18)
plt.show()
```
